Cette question portera sur la création d'un CI fonctionnant dans GitHub. Vous avez évidemment le droit d'utiliser des actions provenant de GitHub pour compléter cette question. Voici les exigences que votre CI devra respecter et le nombre de points alloués pour chacune des exigences :

1. Votre CI devra tourner sur une machine virtuelle Windows.
2. Votre CI devra comprendre une étape qui installera les dépendances qui se situent dans le fichier requirements.txt.
3. Votre CI devra comprendre une étape qui valide Pylint en respectant le fichier de configuration Pylint fourni dans le dossier de cette question qui se nomme pycodestyle.py.
4. Votre CI devra comprendre une étape qui exécutera des tests avec Pylint en respectant le fichier de configuration pytest.ini fourni dans le dossier de cette question.
5. Votre CI devra comprendre une étape qui exécutera la couverture de code avec pytest. Cette étape devra générer un rapport qui devra être automatiquement poussé dans Git dans le dossier de la question 5. Le rapport devra porter le nom suivant : CoverageReport.json. Afin de compléter cette question, vous devez évidemment créer un module de test respectant les exigences de la configuration de pytest et y créer une classe de test. Veuillez nommer le fichier de cette classe de test End_Scan_Tests.py. Elle devra tester toutes les méthodes de la classe EndScan se trouvant dans le fichier End_Scan.py, de manière à avoir une couverture de 100%. Vous serez évalué sur la qualité du code de test. L'architecture et le linter seront également évalués sur cette classe de test.
6. Votre CI devra avoir une étape qui génère un rapport des dépendances. Nous voulons connaître toutes les dépendances de cette solution avec une brève description, le type de licence de chacune des dépendances, la version de chacune des dépendances, un lien vers la licence et/ou un lien vers la dépendance sur le web.
7. Votre CI devra avoir une étape qui nous donnera des recommandations sur le niveau de sécurité de nos dépendances.
8. Votre CI devra mettre le code de cette solution dans un conteneur Docker et générer une image Docker. Nous vous demandons d'ajouter l'image générée dans le répertoire de la question 5 sous le dossier ignoredbypytest. Cet ajout peut être fait manuellement.
