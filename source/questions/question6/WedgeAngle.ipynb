# Finding the angle of a wedge from two echoes.

### Context
The next question requires an understanding of the propagation of ultrasonic waves as well as some basic trigonometry.

Inspectors who use ultrasonic inspection equipment use [wedges](https://www.olympus-ims.com/en/ndt-tutorials/transducers/pa-definitions/) daily to perform their work.

In addition to preserving the proper functioning of the elements that emit and receive ultrasonic signals, they have an essential role in facilitating acoustic coupling. They are particularly useful for orienting beams at the desired angle.

Before beginning an inspection, there are several techniques for calibrating an inspection environment with empirical data (device-probe-wedge-specimen). The question is inspired by a real technique.

### Problem statement

The objective of this question is to determine the angle α (in degrees) of a wedge from the signal generated by two pulses.

![Wedge Angle Image](Images/WedgeAngle.jpg)


![Element 1: 10us](Images/Element1.jpg)
![Element 3: 12us](Images/Element3.jpg)

You are asked to implement the function get_wedge_angle_in_degrees() in the file Wedge_angle.py.

The parameters can be declared and assigned locally.

Do not hesitate to document the approach.

To simplify the problem, it is recommended to consider that the peak of the pulse is centered on the interface that reflects the sound wave. 

The image of the wedge is for illustrative purposes only, dimensions are not to scale.
