### Problem F (1.6)
Que représentent les bytes dans le RawAScan et RawCScan?

A - Pour les AScan c'est sa valeur et pour les CScan c'est la valeur de son status, son amplitude et de sa value.

B - Représente l'adresse mémoire de la valeur qui est convertie en byte par hdf5. Pour les Cscan c'est l'adresse mémoire de son status, son amplitude et de sa value.

C - Le RawAScan a un Data Type de 12-byte tandis que les rawCScan ont un Data Type de 24-byte. Soit 1 byte = 8 bits = 2 hexadécimales. Ainsi, chaque byte correspond à une valeur qui est convertie à sa lecture.

D - Le RawAScan a un Data Type de 12-byte tandis que les rawCScan ont un Data Type de 24-byte. Soit 1 byte = 8 bits = 2 hexadécimales. Ainsi chaque byte correspond à une adresse mémoire d'un objet (opaque) qui permet d'aller lire sa donnée à la lecture.

---

What does the byte represent in the RawAScan and RawCScan?

A - For AScans it is its value and for CScans it is the value of its status, its amplitude and its value.

B - Represents the memory address of the value that is converted to byte by hdf5. For Cscan it is the memory address of its status, its amplitude and its value.

C - The RawAScan has a 12-byte Data Type while the RawCScan has a 24-byte Data Type. Let 1 byte = 8 bits = 2 hexadecimals. Thus each byte corresponds to a value which is converted when it is read.

D - The RawAScan has a 12-byte Data Type while the RawCScans have a 24-byte Data Type. Let 1 byte = 8 bits = 2 hexadecimals. Thus each byte corresponds to a memory address of an object (opaque) which allows its data to be read.
